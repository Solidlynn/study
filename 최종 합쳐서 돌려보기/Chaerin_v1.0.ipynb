## 5. column_name으로 for문을 돌린다

### 5-1. column_name의 list를 만든다

```python
# column_list = ['token_noun', 'token_all', 'token_noun_stopwords', 'token_all_stopwords', 'token', 'token_stopwords']
column_list = ['token_noun_stopwords']
```

```python
nlp, positive_keywords, negative_keywords = initialize_sentiment_analysis(pos_file_path, neg_file_path)
```

```python
for column_name in tqdm(column_list, desc="Processing Columns"):
    # 데이터분석을 한다 - 감성분석을 한다
    df_docu_sent = sentiment_analysis(df_docu_source, column_name, nlp, positive_keywords, negative_keywords)
    df_sent_sent = sentiment_analysis(df_sent_source, column_name, nlp, positive_keywords, negative_keywords)
    # 데이터분석을 한다 - 주제분석을 한다  
    df_topic, df_docu_topic = topic_modeling(df_docu_source, column_name, no_topic)

    # 컬럼별로 엑셀 파일을 분리한다.
    excel_file = './' +  file_name.rsplit('.txt', 1)[0] + '/' + file_name.rsplit('.txt', 1)[0] + '_' + column_name + '.xlsx'

    sheet_name = '주제'
    save_dataframe_to_excel(df_topic, excel_file, sheet_name)

    # 순차분석을 한다. (코치 > 코치이, 코치이 > 코치)
    df_sequence_sent = sequencial_coach_2_coachi(df_docu_sent)
    sheet_name = '코치to코치이_감성' 
    save_dataframe_to_excel(df_sequence_sent, excel_file, sheet_name)
    
    df_sequence_sent_reverse = sequencial_coachi_2_coach(df_docu_sent)
    sheet_name = '코치이to코치_감성' 
    save_dataframe_to_excel(df_sequence_sent_reverse, excel_file, sheet_name)
    
    # 코치-코치이 주제분석을 한다.
    df_sequence_topic = topic_coach_2_coachi(df_docu_topic)
    sheet_name = '코치to코치이_주제' 
    save_dataframe_to_excel(df_sequence_topic, excel_file, sheet_name)
    
    df_sequence_topic_reverse = topic_coachi_2_coach(df_docu_topic)
    sheet_name = '코치이to코치_주제' 
    save_dataframe_to_excel(df_sequence_topic_reverse, excel_file, sheet_name)
    
    # 감정흐름선 분석을 한다.
    df_sentiment_flow = sentiment_flow(df_sent_sent)
    sheet_name = '감정흐름'
    save_dataframe_to_excel(df_sentiment_flow, excel_file, sheet_name)

    # df_sent_sent sheet
    sheet_name = '대화원문'
    save_dataframe_to_excel(df_sent_sent, excel_file, sheet_name)

    # df_talk_portion
    sheet_name = '발언비중'
    save_dataframe_to_excel(df_talk_portion, excel_file, sheet_name)
    
    # View별로 분석을 한다 : talker와 stage 각각의 값과 '전체'를 포함한 리스트 생성
    talkers = df_docu_source['talker'].unique().tolist() + ['전체']
    stages = df_docu_source['stage'].unique().tolist() + ['전체']
    
    # 각 조건에 맞는 행을 추출하여 데이터프레임에 저장
    for talker in talkers[::-1]:
        for stage in stages[::-1]:
            
            view_name = talker + '-' + stage
            # print(view_name, column_name)

            # View 별로 group화한 텍스트에 대한 분석을 한다.
            
            # df_group의 talker, stage에 해당하는 row를 df_source에 저장한다.
            df_source = df_group[(df_group['talker'] == talker) & (df_group['stage'] == stage)]        
            # word_list 생성한다.
            word_list = extract_wordlist(df_source, column_name)
            
            # 단어 빈도 분석을 한다.
            df_top_f_words = get_top_word_frequencies(word_list, no_top_frequency)
            sheet_name = view_name + '-' + '빈도' 
            save_dataframe_to_excel(df_top_f_words, excel_file, sheet_name)
            
            # 중심단어 분석을 한다.
            df_top_n_words, G = extract_top_words_and_save_graph(word_list, view_name, column_name, font_path, no_top_network)
            # 중심단어별 연관어 분석을 한다.
            df_top_n_r_words = extract_top_related_words(G, df_top_n_words, no_top_network)
            sheet_name = view_name + '-' + '중심'
            save_dataframe_to_excel(df_top_n_r_words, excel_file, sheet_name)            
            
            # 감성분석을 한다.
            df_view_sent = calculate_mean_sent_score(df_docu_source, talker, stage)
            sheet_name = view_name + '-' + '감성'
            save_dataframe_to_excel(df_view_sent, excel_file, sheet_name)
            
            # 빈도 top-words가 들어간 문장의 감성지수를 분석한다.
            df_top_f_sent = analyze_top_words_sent(df_top_f_words, df_sent_source, talker, stage, column_name, view_name)
            sheet_name = view_name + '-' + '빈도_감성' 
            save_dataframe_to_excel(df_top_f_sent, excel_file, sheet_name)

            # 네트워크 top-words가 들어간 문장의 감성지수를 분석한다.
            df_top_n_sent = analyze_top_words_sent(df_top_n_words, df_sent_source, talker, stage, column_name, view_name)
            sheet_name = view_name + '-' + '중심_감성' 
            save_dataframe_to_excel(df_top_n_sent, excel_file, sheet_name)
```
